```python
# required library for data analysis
import pandas as pd
import numpy as np

# required librairies for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('talk')
```

```python

# import the data 
# reading the csv file
df = pd.read_csv("telecom_data.csv")

# let's look inside the data
display(df.head())
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,12))

plt.subplot(2,3,1)
sns.histplot(data=df, x='Total_calls')
plt.title("row = 1, column = 1, position = 1")

plt.subplot(2,3,2)
sns.histplot(data=df, x='Total_calls', binwidth=50)
plt.title("row = 1, column = 2, position = 2")

plt.subplot(2,3,3)
sns.histplot(data=df, x='Total_calls', binwidth=20)
plt.title("row = 1, column = 3, position = 3")

plt.subplot(2,3,4)
sns.histplot(data=df, x='Total_calls', binwidth=20)
plt.xticks(np.arange(191, 417, 20), rotation=90)
plt.title("row = 2, column = 1, position = 4")

plt.subplot(2,3,5)
sns.histplot(data=df, x='Total_calls', binwidth=20, color='orange')
plt.title("row = 2, column = 2, position = 5")

plt.subplot(2,3,6)
sns.histplot(data=df, x='Total_calls', binwidth=20, color='orange', fill=False)
plt.title("row = 2, column = 3, position = 6")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,12))

plt.subplot(2,3,1)
sns.histplot(data=df, x='Total_calls', binwidth=1)
plt.title("row = 1, column = 1, position = 1")

plt.subplot(2,3,2)
sns.histplot(data=df, x='Total_calls', binwidth=1, cumulative=True)
plt.grid()
plt.title("row = 1, column = 2, position = 2")

plt.subplot(2,3,3)
sns.histplot(data=df, x='Total_calls', binwidth=1, cumulative=True, stat='probability')
plt.grid()
plt.title("row = 1, column = 3, position = 3")

plt.subplot(2,3,4)
sns.histplot(data=df, x='Total_calls', binwidth=1, kde=True)
plt.title("row = 2, column = 1, position = 4")

plt.subplot(2,3,5)
sns.histplot(data=df, x='Total_calls', binwidth=1, kde=True, hue='Churn')
plt.title("row = 2, column = 2, position = 5")

plt.subplot(2,3,6)
sns.histplot(data=df, x='Total_calls', binwidth=1, cumulative=True, stat='probability', hue='Churn')
plt.grid()
plt.title("row = 2, column = 3, position = 6")

plt.tight_layout()
plt.show()
```
