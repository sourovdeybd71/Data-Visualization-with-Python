```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('talk')
```

```python
df = pd.read_csv("bmw.csv")
display(df.head())
```

# We use scatterplot for numerical Variables, and barplot for categorical variables. 

#Find the relation between the numerical variables
Scatterplot of mileage vs price
Scatterplot of mpg vs price
Scatterplot of engineSize vs price

```python
plt.figure(figsize=(20,12))
#milage vs price
plt.subplot(3,1,1)
sns.scatterplot(x='mileage', y='price', data=df)

#mpg vs price
plt.subplot(3,1,2)
sns.scatterplot(x='mpg', y='price', data=df)

# engineSize vs price
plt.subplot(3,1,3)
sns.scatterplot(x='engineSize', y='price', data=df)

plt.tight_layout()
plt.show()
```

Uses of "hue"

```python
plt.figure(figsize=(20,12))

#mileage vs price 
plt.subplot(2,2,1)
sns.scatterplot(x='mileage', y='price', data=df)

#adding a categorical feature: "transmission"
plt.subplot(2,2,2)
sns.scatterplot(x='mileage', y='price', data=df, hue='transmission')

#adding a categorical feature "fueltype"
plt.subplot(2,2,3)
sns.scatterplot(x='mileage', y='price', data=df, hue='fuelType')

plt.tight_layout()
plt.show()
```
