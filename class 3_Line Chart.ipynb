```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
m=1
c=1
SL= pd.DataFrame()
SL['x'] = np.arange(-10,10,1)

SL["y"]=m*SL["x"]+c
display(SL)
```

```python
plt.figure(figsize=(12,5))
plt.plot(SL['x'],SL["y"], marker="o")
plt.vlines(0,-10,15, linestyles=":", color="black")
plt.hlines(0,-10,15, linestyles=":",color="black")

plt.title("y=mx+c where m=1, and c=1")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-10,11,2))
plt.yticks(np.arange(-10,15,2))

plt.show()
```

```python
def straight_line(m,c):
    df= pd.DataFrame()
    df['x'] = np.arange(-10,10,1)

    df["y"]=m*SL["x"]+c
    return df

SL_1= straight_line(1,1)
SL_2= straight_line(1,-1)
SL_3= straight_line(-1,1)
SL_4= straight_line(-1,-1)
SL_5= straight_line(1,0)
SL_6= straight_line(-1,0)
display(SL_1)
```

```python
sns.set_context("talk")


def line_plot(x_col, y_col, title):
    
    plt.plot(x_col,y_col, marker="o")
    plt.vlines(0,-15,20, linestyles=":", color="black")
    plt.hlines(0,-10,10, linestyles=":",color="black")

    plt.title("y=mx+c where "+title)
    plt.xlabel("x")
    plt.ylabel("y")
#     plt.xticks(np.arange(-10,11,2))
#     plt.yticks(np.arange(-10,15,2))

plt.figure(figsize=(20,8))
plt.subplot(2,3,1)
line_plot(SL_1["x"], SL_1["y"], 'm=1 c=1')
plt.subplot(2,3,2)
line_plot(SL_2["x"], SL_2["y"], 'm=1 c=-1')
plt.subplot(2,3,3)
line_plot(SL_3["x"], SL_3["y"], 'm=-1 c=1')
plt.subplot(2,3,4)
line_plot(SL_4["x"], SL_4["y"], 'm=-1 c=-1')
plt.subplot(2,3,5)
line_plot(SL_5["x"], SL_5["y"], 'm=1 c=0')
plt.subplot(2,3,6)
line_plot(SL_6["x"], SL_6["y"], 'm=-1 c=1')

plt.tight_layout()
plt.show()
```

```python
#multiple line in same plot

plt.figure(figsize=(20,12))

plt.plot(SL_1['x'], SL_1['y'], color= 'red', label='m=1 c=1')
plt.plot(SL_2['x'], SL_2['y'], color= 'green', label='m=1 c=-1')
plt.plot(SL_3['x'], SL_3['y'], color= 'blue', label='m=-1 c=1')
plt.plot(SL_4['x'], SL_4['y'], color= 'orange', label='m=-1 c=-1')
plt.plot(SL_6['x'], SL_5['y'], color= 'yellow', label='m=1 c=0')
plt.plot(SL_6['x'], SL_6['y'], color= 'purple', label='m=-1 c=0')

plt.vlines(0, -15, 15, linestyles=':', colors="black")
plt.hlines(0, -10, 10, linestyles=':', colors="black")
plt.title("y=mx+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
